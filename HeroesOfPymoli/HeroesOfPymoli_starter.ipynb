```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchasedata_df = pd.read_csv(file_to_load)
```

```python
purchasedata_df.head()
```

```python
# Display total number of players
sn_count = len(purchasedata_df["SN"].unique())
player_count = pd.DataFrame({"Total Players": [sn_count]})
player_count
```

```python
#Purchasing Analysis
#Optain unique items, average price, etc.
#Create a summary data frame for results
#Display new data frame

item_name_count = len(purchasedata_df["Item Name"].unique())
average_price = purchasedata_df["Price"].mean()
number_of_purchases = purchasedata_df["Purchase ID"].count()
total_revenue = sum(purchasedata_df["Price"])

purchase_analysis_df = pd.DataFrame({"Number of Unique Items": [item_name_count], "Average Price": [average_price],
                                     "Number of Purchases": [number_of_purchases], "Total Revenue": [total_revenue]})
purchase_analysis_df
```

```python
#Gender demographics
#Percentage & Count of Males
#Percentage & Count of females
#Percentage & Count of Other
gender_duplicates_df = purchasedata_df.loc[:,["Gender", "SN", "Age"]]
gender_duplicates_df = gender_duplicates_df.drop_duplicates()
gender_count = gender_duplicates_df["Gender"].value_counts()
# # # gender_duplicates_df = pd.DataFrame(purchasedata_df.drop_duplicates(["SN"]))
# # # # gender_grouping = gender_duplicates_df.groupby(["Gender"]).count()["Gender"]
# # # gender_duplicates_df
player_per = gender_count / sn_count * 100
gender_demographics = pd.DataFrame({"Total Count": gender_count, "Percentage of Players": player_per})
# gender_demo_df = pd.DataFrame({"Total Count": [gender_count], "Percentage of Players": [player_per]})
# # player_per
gender_demographics
# total_players
```

```python
#Purchasing Analysis (Gender)
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
gender_count = purchasedata_df.groupby(["Gender"]).count()["Purchase ID"]
gender_average = purchasedata_df.groupby(["Gender"]).mean()["Price"]
gender_purchase_value = purchasedata_df.groupby(["Gender"]).sum()["Price"]
gender_avg_per_person = gender_purchase_value / gender_demographics["Total Count"]

# # Create a summary data frame to hold the results
analysis_gender = pd.DataFrame({"Purchase Count": gender_count, "Average Purchase Price": gender_average, 
                                "Total Purchase Value": gender_purchase_value, 
                                "Avg Total Purchase per Person": gender_avg_per_person})

# # Optional: give the displayed data cleaner formatting
# # Display the summary data frame
analysis_gender
```

```python
# Age Demographics
# Establish bins for ages
# Categorize the existing players using the age bins. Hint: use pd.cut()
# Calculate the numbers and percentages by age group
# Create a summary data frame to hold the results
# Optional: round the percentage column to two decimal points
# Display Age Demographics Table

group_ranges = ['<10', '10-14' ,'15-19', '20-24', '25-29', '30-34', '35-39', '40+']
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 1000]
gender_duplicates_df["Age Range"] = pd.cut(gender_duplicates_df["Age"], bins, labels= group_ranges)

age_count = gender_duplicates_df["Age Range"].value_counts()
per_count = age_count / sn_count * 100

age_demographics = pd.DataFrame({"Total Count": age_count, "Percentage of Players": per_count})
age_demographics
```

```python
# Purchasing Analysis (Age)
# Bin the purchase_data data frame by age
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame

purchasedata_df["Age Range"] = pd.cut(purchasedata_df["Age"], bins, labels=group_ranges)


age_count = purchasedata_df.groupby(["Age Range"]).count()["Purchase ID"]
age_average = purchasedata_df.groupby(["Age Range"]).mean()["Price"]
age_purchase_value = purchasedata_df.groupby(["Age Range"]).sum()["Price"]
age_avg_per_person = age_purchase_value / age_demographics["Total Count"]

# # Create a summary data frame to hold the results
analysis_age = pd.DataFrame({"Purchase Count": age_count, "Average Purchase Price": age_average, 
                                "Total Purchase Value": age_purchase_value, 
                                "Avg Total Purchase per Person": age_avg_per_person})
analysis_age
```

```python
# Top Spenders
# Run basic calculations to obtain the results in the table below
# Create a summary data frame to hold the results
# Sort the total purchase value column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame

player_count = purchasedata_df.groupby(["SN"]).count()["Price"]
player_average = purchasedata_df.groupby(["SN"]).mean()["Price"]
player_total = purchasedata_df.groupby(["SN"]).sum()["Price"]

spender_summary=pd.DataFrame({"Purchase Count":player_count,"Average Purchase":player_average,
                          "Total Purchase Value":player_total})

spender_summary.sort_values("Total Purchase Value",ascending=False).round(2).head()
```

```python
# Most Popular Items
# Retrieve the Item ID, Item Name, and Item Price columns
# Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
# Create a summary data frame to hold the results
# Sort the purchase count column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame

item_id = purchasedata_df.groupby(["Item ID"])
item_name = purchasedata_df.groupby(["Item Name"]).unique()
item_purchase_count = item_id["SN"].count()
item_price = item_id["Price"].unique()
item_purchase_value = item_id["Price"].sum()

item_summary = pd.DataFrame({"Item ID":item_id,"Item Name":item_name,"Purchase Count":item_purchase_count,
                             "Item Price":item_price,"Total Purchase Value":item_purchase_value})

item_summary = item_summary.sort_values('Purchase Count', ascending=False)
item_summary_df = item_summary[['Item Name','Purchase Count','Item Price','Total Purchase Value']]

item_summary_df.head()
```

```python
# Most Profitable Items
# Sort the above table by total purchase value in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the data frame

most_profitable = item_summary.sort_values('Total Purchase', ascending=False)
most_profitable.head()
```
